# Question 3
#### What is the fuel type with lowest average fuel cost per unit burned?

```python
# group the average fuel cost per unit burned by fuel type
avg_cost_burned = data.groupby("fuel_type_code_pudl", as_index=False)["fuel_cost_per_unit_burned"].mean()
avg_cost_burned
```

```python
# sort values from lowest by fuel cost per unit burned 
sorted_avg_cost_burned = avg_cost_burned.sort_values(by= "fuel_cost_per_unit_burned")
sorted_avg_cost_burned
```

```python
# plot the average fuel cost per unit burned by fuel type in descending order
# exclude the obvious nuclear fuel type
sns.barplot(data=sorted_avg_cost_burned[:-1], x= "fuel_type_code_pudl", y= "fuel_cost_per_unit_burned")
plt.plot()
```

```python
least_avg_cost_burned = sorted_avg_cost_burned.iloc[0]["fuel_type_code_pudl"]
print(f'The fuel type with the lowest average fuel cost per unit burned is: {least_avg_cost_burned}')
```

# Question 4

#### What is the standard deviation and 75th percentile of the measure ofenergy per unit (Fuel_mmbtu_per_unit) in two decimal places?

```python
fuel_mmbtu_std = data["fuel_mmbtu_per_unit"].std()
print('The standard deviation  of the measure of energy per unit: {0:.2f}'.format(fuel_mmbtu_std))
```

# Question 9

#### For the fuel type coal, what is the percentage change in the fuel cost per unit burned in 1998 compared to 1994?

```python
# group the sum of fuel cost per unit burned by fuel type and report year
group = data.groupby(["fuel_type_code_pudl", "report_year"])["fuel_cost_per_unit_burned"].sum()
pd.DataFrame(group)
```

```python
# get the total Cost of Coal per unit burned for each year
coal_cost = pd.DataFrame(group.loc['coal'].rename("Cost of Coal per unit burned"))
coal_cost
```

```python
coal_cost.plot(kind='bar')
plt.plot()
```

```python
# locate the years 1994 and 1998
coal_cost.loc[[1994, 1998]]
```

```python
coal_cost.loc[[1994, 1998]].plot(kind="bar")
plt.plot()
```

```python
# get the total cost of Coal per unit burned for 1994 
cost_1994 =coal_cost.loc[1994].values[0]
cost_1994
```

```python
# get the total cost of Coal per unit burned for 1998 
cost_1998 =coal_cost.loc[1998].values[0]
cost_1998
```

```python
percent_change = ((cost_1994 - cost_1998)/cost_1994) * 100
print(f'percentage change in the fuel cost per unit burned in 1998 compared to 1994 is: {percent_change.round()}%')
```

# Question 10

#### Which year has the highest average fuel cost per unit delivered?

```python
# group the average fuel cost per unit delivered by the report year
delivered_per_year = data.groupby("report_year", as_index= False)["fuel_cost_per_unit_delivered"].mean()
delivered_per_year
```

```python
plt.figure(figsize= (15, 6))
sns.barplot(data= delivered_per_year, x= "report_year", y= "fuel_cost_per_unit_delivered" )
plt.yticks(rotation=60)
plt.plot()
```

```python
# sort the fuel_cost_per_unit_delivered by descending order
delivered_per_year.sort_values(ascending=False, by= "fuel_cost_per_unit_delivered")
```

```python
# get the row with the maximum fuel_cost_per_unit_delivered  
highest_del = delivered_per_year[delivered_per_year["fuel_cost_per_unit_delivered"]== delivered_per_year["fuel_cost_per_unit_delivered"].max()]
highest_del
```

```python
print(f'Year when highest average fuel cost per unit was delivered: {highest_del["report_year"].values[0]}')
```
